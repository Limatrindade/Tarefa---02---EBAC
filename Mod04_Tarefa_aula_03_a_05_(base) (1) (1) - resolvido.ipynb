## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui

valor_um_porcento_mat = 100 / 24
acerto_percentual_mat = round(valor_um_porcento_mat * df_mat)
acerto_percentual_mat

valor_um_porcento_por = 100 / 18
acerto_percentual_por = round(valor_um_porcento_por * df_por)
acerto_percentual_por

valor_um_porcento_geo = 100 / 8
acerto_percentual_geo = round(valor_um_porcento_geo * df_geo)
acerto_percentual_geo

valor_um_porcento_ing = 100 / 8
acerto_percentual_ing = round(valor_um_porcento_ing * df_ing)
acerto_percentual_ing

valor_um_porcento_his = 100 / 8
acerto_percentual_his = round(valor_um_porcento_his * df_his)
acerto_percentual_his

valor_um_porcento_fis = 100 / 12
acerto_percentual_fis = round(valor_um_porcento_fis * df_fis)
acerto_percentual_fis

valor_um_porcento_qui = 100 / 12
acerto_percentual_qui = round(valor_um_porcento_qui * df_qui)
acerto_percentual_qui
````

````python
# 2) Seu código aqui

total_acertos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
total_acertos['Total de Acertos'] = total_acertos.sum(axis=1)
print(total_acertos)
````

````python
# 3) Seu código aqui

total_questoes = 24 + 18 + 8 + 8 + 8 + 12 + 12
percentual_geral = (total_acertos['Total de Acertos'] / total_questoes) * 100
total_acertos['Percentual Geral'] = percentual_geral
print(total_acertos)
````

````python
# 4) Seu código aqui

nota_de_corte = 45
alunos_aprovados = total_acertos[total_acertos['Percentual Geral'] > nota_de_corte]
quantidade_alunos_aprovados = len(alunos_aprovados)
print("Número de alunos com nota maior que 45:", quantidade_alunos_aprovados)
````
